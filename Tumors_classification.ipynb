```python
print("Hello World.")


from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


dataset = load_breast_cancer()#Loading the dataset.
print(dataset)


print("The shape of data:",dataset.data.shape)
print("The description of data:",dataset.DESCR)
print(dataset.keys())

#Creating a pandas dataframe.
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)

df['target'] = dataset.target#The target data, which contains the predictions made.
#It is the variable predicted. In this project, is is analysed if the tumor is malignant or benign.

print(df.describe())
print("Missing values:",df.isnull().sum())
print(df.head())

X = dataset.data#The data used in the project.
y = dataset.target#The target data.


#Split th data into train data and test data.
X_train,X_test,y_train,y_test = train_test_split(X,y)

classification = KNeighborsClassifier(n_neighbors=8)
classification.fit(X_train,y_train)
classification_score = classification.score(X_test,y_test)
print(classification_score*100)



```
